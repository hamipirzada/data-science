 ## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting  in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width and height of figure

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="X-axis",
      ylabel="Y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

### Two options for subplots


```python
# Subplotsoption 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2,
                                             ncols= 2,
                                             figsize= (10, 5))
# Plot to different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5));

# Plot to different axis 
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot or matplotlib oop method?)
* When plotting something quickly, okay to use the pyplot method.
* When plotting something advanced, use the oop method.

```python
heart_disease
```

```python
# Patients above 50 age
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot method.
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             figsize=(10, 6));
```

```python
# OOP method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# ax.set_xlim([45, 100]);
```

```python
heart_disease
```

```python
# OOP method
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="summer")
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
heart_disease
```

```python
# Subplot  of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
# customize a0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
# add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
# Customize ax1
ax1.set(title="Heart disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
# add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use("seaborn-whitegrid")
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")    # this changes the colour scheme
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");
# add a horizontal mean line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Customizing the x and y axis limitations
# Subplot  of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")
# customize a0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
# change the x axis limits
ax0.set_xlim([50, 80])
# change the y axis limits
ax0.set_ylim([50, 400])
# add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="summer")
# Customize ax1
ax1.set(title="Heart disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
# change the y axis limits
ax1.set_ylim([60, 200])
# add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
